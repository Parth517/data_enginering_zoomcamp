```python
df_green_trip["lpep_pickup_datetime"] = pd.to_datetime(df_green_trip["lpep_pickup_datetime"])

# Filter trips with distance < 100 miles
filtered_df = df_green_trip[df_green_trip["trip_distance"] < 100]

# Extract pickup day (date only)
filtered_df["pickup_day"] = filtered_df["lpep_pickup_datetime"].dt.date

# Find the pickup day with the longest trip distance
result = (
    filtered_df
    .loc[filtered_df["trip_distance"].idxmax(), "pickup_day"]
)

result
```

```python
df_nov18 = df_green_trip[
    (df_green_trip["lpep_pickup_datetime"] >= "2025-11-18") &
    (df_green_trip["lpep_pickup_datetime"] < "2025-11-19")
]
```

```python
total_by_location = (
    df_nov18
    .groupby("PULocationID")["total_amount"]
    .sum()
    .reset_index()
)
```

```python
result = (
    total_by_location
    .merge(df_zone_data, left_on="PULocationID", right_on="LocationID", how="left")
    .sort_values("total_amount", ascending=False)
    .head(1)
)
result
```

```python
result = (
    df_green_trip.assign(lpep_pickup_datetime=pd.to_datetime(df_green_trip["lpep_pickup_datetime"]))
      .merge(
          pd.read_csv("taxi_zone_lookup.csv")[["LocationID", "Zone"]],
          left_on="PULocationID",
          right_on="LocationID",
          how="left"
      )
      .rename(columns={"Zone": "pickup_zone"})
      .query(
          "pickup_zone == 'East Harlem North' "
          "and lpep_pickup_datetime >= '2025-11-01' "
          "and lpep_pickup_datetime < '2025-12-01'"
      )
      .merge(
          pd.read_csv("taxi_zone_lookup.csv")[["LocationID", "Zone"]],
          left_on="DOLocationID",
          right_on="LocationID",
          how="left"
      )
      .rename(columns={"Zone": "dropoff_zone"})
      .groupby("dropoff_zone", as_index=False)["tip_amount"]
      .max()
      .sort_values("tip_amount", ascending=False)
      .head(1)
)

result
```
